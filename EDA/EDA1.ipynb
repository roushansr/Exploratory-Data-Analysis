```python
#df.corr() here takes the String values also. Hence gives and output error. So select only int and float data.

dfV1 = df.select_dtypes(include=[float, int]).corr()
```

```python
sns.heatmap(dfV1, annot = True)

# annot gives the quick numbers

# Makes the pot looks big and gives a cleaner look
plt.rcParams['figure.figsize'] = (25,10)
plt.show()
```
