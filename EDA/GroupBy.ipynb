```python
import pandas as pd
```

```python
df = pd.read_csv('Flavors.csv')
df
```

```python
df.nunique()

#get the unique values, so you know which can be grouped. The more the unique values, the better to apply the GroupBy function.
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'], 'Texture Rating': ['mean','max','count','sum']})
```
