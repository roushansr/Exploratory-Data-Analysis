#### This can be an easy way to remove characters from a column having string values.
#### This uses the 'Regular Expression' library.

```python
import re
df1 = pd.read_excel(r"Customer Call List.xlsx")
df1['Last_Name'] = df1['Last_Name'].apply(lambda x: re.sub('[^a-zA-Z]', '', x)
                                          if pd.notna(x) else x)
df1
```

```python
# Replacing the characters in Phone number.
import string
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]','', regex = True)

#String library is used with replace() function. Replace [^0-9] with Nothing i.e. replaces everything except 0-9 with Nothing.

df["Phone_Number"] = df["Phone_Number"].str.replace('NaN', '', regex = True) #this didn't work.
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].fillna('')

#Using this 'fillna()' function, replaced the unnecessary data(NaN).
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y', regex = True)
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N', regex = True)
df["Paying Customer"] = df["Paying Customer"].fillna('')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y', regex = True)
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N', regex = True)
df
```

## We can replace N/a, but we have to use fillna() function for NaN.

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','', regex = True)
df
```

```python
#We perform the same across the dataframe.

df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna('')
df["Last_Name"] = df["Last_Name"].fillna('')
df["Address"] = df["Address"].str.replace('N/a','', regex = True)
```
